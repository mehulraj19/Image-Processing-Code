Frequency Domain Tasks in OpenCV

Two tasks:
<ul>

<li>Image Sharpness</li>

```
# Importing packages
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

# Low Frequency Filters
<br/>
Low Frequency filters help us to give smoothing to images. One can see and differentiate this by seeing smooth edges.  

```
# reading image
img = cv.imread('image1.jpg')
```

```
#displaying image
cv2_imshow(img)
```

```
# prepare 3x3 shaped filter
kernel = np.array([[1,1,1],[1,1,1],[1,1,1]])
#print(np.sum(kernel))
kernel = kernel/sum(kernel)
print(kernel)
```

```
# filter source image
res = cv.filter2D(img, -1, kernel);
```

```
# Saving result image
cv.imwrite('result.jpg', res)
```

```
# displaying iamge
cv2_imshow(res)
```

```
blur = cv.GaussianBlur(img, (5,5),0)
```

```
cv2_imshow(blur)
```

# High Frequency Filters
<br/>These filters help us to increase the sharpness of images.

```
# Reading Image
img = cv.imread('image2.jpg')
```

```
# Displaying Image
cv2_imshow(img)
```

```
# kernel setup
kernel = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
print(np.sum(kernel))
kernel = kernel / (np.sum(kernel) if np.sum(kernel)!=0 else 1)
print(kernel) 
```

```
# filter the source image
res = cv.filter2D(img,-1,kernel)
```

```
# saving result image
cv.imwrite('result1.jpg',res)
```

```
# displaying iamge
cv2_imshow(res)
```

```
blur = 1 - cv.GaussianBlur(img,(5,5),0)
cv2_imshow(blur)
```
