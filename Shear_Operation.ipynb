Shear Operation on Images

Importing packages

```
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
```

Getting image

```
image_file1 = "/content/image1.jpg"
```

Printing Image

```
img1 = cv.imread(image_file1,cv.IMREAD_UNCHANGED)
cv2_imshow(img1)
```

Now important note is that the Shear Operation is a Geometric Operation and this all happens using an array multiplication in broad aspect, will make an array first  

```
m = np.array([[1,-0.1,0],[0,1,0]]) 
# This will affect the image by shearing this to left direction, will be more clear when see in action
```

```
m = np.float32(m)
#This is just to make the values of the array to particular format 
```

```
rows,col,ch = img1.shape
#This will help us to get the tuple of values that actually shapes the image
```

```
dst = cv.warpAffine(img1, m, (rows,col)) 
#This wiil affect the image or img1 and will help us to do the neccessary action 
```

```
cv2_imshow(dst)
#One can clearly see the shear!!
```

Now will do the same task but the right shift

```
m = np.array([[1,0.1,0],[0,1,0]]) 
#This matricx will help us to achieve the shear
```

```
m = np.float32(m)
#This will help us to get the values in particular format 
```

```
dst2 = cv.warpAffine(img1, m, (rows,col))
cv2_imshow(dst2)
```

We can apply many affects to our image by simply making changes in the 'm' matrix. Will perform some more actions!!

```
m = np.array([[1,0,10],[0,1,0]]) 
m = np.float32(m)
dst = cv.warpAffine(img1, m, (rows,col))
cv2_imshow(dst)
#Shifting the whole image as it is to right
```

```
m = np.array([[1,0,0],[0,1,10]]) 
m = np.float32(m)
dst = cv.warpAffine(img1, m, (rows,col))
cv2_imshow(dst)
#Shifting the image down
```

```
m = np.array([[1,0,0],[0.8,1,0]]) 
m = np.float32(m)
dst = cv.warpAffine(img1, m, (rows,col))
cv2_imshow(dst)
#Shifting image towards us to left position
```

```
m = np.array([[0.8,0,0],[0,1,0]]) 
m = np.float32(m)
dst = cv.warpAffine(img1, m, (rows,col))
cv2_imshow(dst)
#Shifting image to left as it is!!
```

```
m = np.array([[1,0,0],[0,0.8,0]]) 
m = np.float32(m)
dst = cv.warpAffine(img1, m, (rows,col))
cv2_imshow(dst)
#Shifting iamge upwards
```

```
m = np.array([[1,0,0],[-0.8,1,0]]) 
m = np.float32(m)
dst = cv.warpAffine(img1, m, (rows,col))
cv2_imshow(dst)
#Shifting image like we are standing on roof and seeing down
```
